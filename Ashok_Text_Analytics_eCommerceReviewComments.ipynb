```python
#convert to list
data = df['Review Text'].tolist()
type(data)
```

```python
#create word list for each token
word_list =[]
#i=0
for record in data:
    #print(record)
    #i = i+1
    #print(i)
    words = []
    tokens = record.split()
    #print(tokens)
    for token in tokens:
        words.append(token.lower())
    word_list.append(words)
word_list[0:10]
```

```python
lex_file = open('D:\\01_Reva\\PDGM\\12_Text_Analytics\\AFINN-111.txt')
```

```python
lexicons = {}
records = lex_file.readlines()
```

```python
records[0]
```

```python
for record in records:
    lexicons[record.rstrip('\n').split(',')[0]] = int(record.rstrip('\n').split(',')[1])
```

```python
#For each tweet
#For each word in tweet
#Get the word score (score is a number if the word is in Lexicon, 0 if not)
#Add all the scores and find the ploarity
```

```python
#check strength of each tweet
```

```python
strength = []
for tweet in word_list:
    #print(tweet)
    score = 0
    for word in tweet:
        if word in (lexicons):
            score = score + lexicons[word]
    strength.append(score)
```
